```python
movies = movies[['title','genres','keywords','overview']]
```

```python
df = movies.merge(credit)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
import json
```

```python
df['genres'] = df['genres'].apply(lambda x: [d['name'] for d in json.loads(x)])
```

```python
df['keywords'] = df['keywords'].apply(lambda x:[d['name'] for d in json.loads(x)])
```

```python
# Function to extract first 3 names safely
def extract_cast_names(text):
    parsed = json.loads(text)
    return [item['name'] for item in parsed[:3] if 'name' in item]
```

```python
df['cast'] = df['cast'].apply(extract_cast_names)
```

```python
def extract_director(text):
    parsed = json.loads(text)
    job = [item['name'] for item in parsed if item.get('job') == 'Director']
    return job
```

```python
df['crew'] = df['crew'].apply(extract_director)
```

```python
df['overview'] = df['overview'].apply(lambda x : x.split())
```

```python
df.head()
```

```python
def remove_spaces(obj):
    return [name.replace(' ','') for name in obj]
```

```python
df['genres'] = df['genres'].apply(remove_spaces)
```

```python
df['keywords'] = df['keywords'].apply(remove_spaces)
```

```python
df['cast'] = df['cast'].apply(remove_spaces)
```

```python
df['crew'] = df['crew'].apply(remove_spaces)
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head()
```

```python
newdf = df[['movie_id','title','tags']]
```

```python
newdf['tags'] = newdf['tags'].apply(lambda x:" ".join(x))
```

```python
newdf['tags']
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
def transform_text(text):
    text = text.lower()
    tokens = nltk.word_tokenize(text)

    tokens = [t for t in tokens if t.isalnum()]
    tokens = [t for t in tokens if t not in stopwords.words('english')]

    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t,pos='v') for t in tokens]
    
    return " ".join(tokens)
```

```python
newdf['tags'] = newdf['tags'].apply(transform_text)
```

```python
newdf
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=3500)
```

```python
vectors = tfidf.fit_transform(newdf['tags']).toarray()
```

```python
len(tfidf.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])
```

```python
def recommend(movie):
    movie_index = newdf[newdf['title'] == movie].index[0]
    top_movies = sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x: x[1])[1:6]
    for i in top_movies:
        print(newdf.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
